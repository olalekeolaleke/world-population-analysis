```python
# Importing Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

# Extract - Store CSV into DataFrame

```python
# Creating csv path and loading first data in Pandas DataFrame
csv_path = ("Resources/World-population-by-countries-dataset.csv")
wrld_pop_df = pd.read_csv(csv_path)
wrld_pop_df.head()
```

```python
# Creating csv path and loading first data in Pandas DataFrame
csv_path = ("Resources/unemployment analysis.csv")
unemp_rate_df = pd.read_csv(csv_path)
unemp_rate_df.head()
```

# Data Cleaning and Transformation 

```python
# Creating a filtered dataframe from specific columns

wrld_pop_cols = ['Country Name', '2011', '2012', '2013', '2014', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

wrld_pop_transformed = wrld_pop_df[wrld_pop_cols].copy()

# Clean the data by dropping duplicates and setting the index
wrld_pop_transformed.drop_duplicates()

# Setting the Country Name column as index
wrld_pop_transformed.set_index("Country Name", inplace=True)

wrld_pop_transformed.head()
```

```python
# Creating a filtered dataframe from specific columns

unemp_rate_cols = ['Country Name', '2011', '2012', '2013', '2014', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

unemp_rate_transformed = unemp_rate_df[unemp_rate_cols].copy()

# Clean the data by dropping duplicates and setting the index
unemp_rate_transformed.drop_duplicates()

# Setting the Country Name column as index
unemp_rate_transformed.set_index("Country Name", inplace=True)

unemp_rate_transformed.head()
```
